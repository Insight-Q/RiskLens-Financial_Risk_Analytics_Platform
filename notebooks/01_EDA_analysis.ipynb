```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/cs-training.csv')
print("Shape:" , df.shape)
df.head()
```

```python
col_info = {
    'SeriousDlqin2yrs': 'TARGET — 1 = defaulted, 0 = did not default',
    'RevolvingUtilizationOfUnsecuredLines': 'Credit card usage ratio (0 to 1)',
    'age': 'Customer age in years',
    'NumberOfTime30-59DaysPastDueNotWorse': 'Times payment was 30-59 days late',
    'DebtRatio': 'Monthly debt payments / Monthly income',
    'MonthlyIncome': 'Monthly income in USD',
    'NumberOfOpenCreditLinesAndLoans': 'Total open loans and credit lines',
    'NumberOfTimes90DaysLate': 'Times payment was 90+ days late',
    'NumberRealEstateLoansOrLines': 'Number of property/mortgage loans',
    'NumberOfTime60-89DaysPastDueNotWorse': 'Times payment was 60-89 days late',
    'NumberOfDependents': 'Number of dependents excluding self'
}

for col, meaning in col_info.items():
    print(f"{col:45} -> {meaning}")
```

```python
print("Missing values before cleaning: ")
print(df.isnull().sum())
print("\nMissing percentange: ")
print((df.isnull().sum() / len(df) * 100).round(2))
```

```python
#fill with median
df['MonthlyIncome'] = df['MonthlyIncome'].fillna(df['MonthlyIncome'].median())
#fill with 0
df['NumberOfDependents'] = df['NumberOfDependents'].fillna(0)

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
#removing outliers
print("Shape before removing outliers:", df.shape)
df = df[(df['age'] > 18) & (df['age'] < 100)]
df = df[df['MonthlyIncome'] < 500000]
df = df[df['DebtRatio'] < 5]

print("Shape after removing outliers:", df.shape)
```

```python
#Feature Engineering
df['age_group'] = pd.cut(df['age'],
                         bins = [18, 30, 45, 60, 100],
                         labels = ['Young', 'Middle', 'Senior', 'Old'])
df['income_tier'] = pd.cut(df['MonthlyIncome'],
                           bins = [0, 3000, 7000, 15000, 50000],
                           labels = ['Low', 'Medium', 'High', 'Very High'])
df['debt_tier'] = pd.cut(df['DebtRatio'],
                         bins = [0, 0.3, 0.6, 1, 5],
                         labels = ['Low Risk', 'Medium Risk', 'High Risk', 'Very High Risk'])
print("New Columns Created Successfully!")
print(df[['age_group', 'income_tier', 'debt_tier']].head(10))
```

```python
plt.figure(figsize=(10,5))
default_by_age = df.groupby('age_group',observed=False)['SeriousDlqin2yrs'].mean()*100
bars = plt.bar(default_by_age.index, default_by_age.values,
               color=['#7F77DD', '#534AB7', '#3C3489', '#26215C'],
               edgecolor = 'white', linewidth=0.5)
for bar,val in zip(bars, default_by_age.values):
    plt.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.2,
             f'{val:.1f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')
plt.title('Default Rate by Age Group', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Age Group', fontsize=12)
plt.ylabel('Default Rate (%)', fontsize=12)
plt.ylim(0, default_by_age.max() + 3)
plt.tight_layout()
plt.savefig('../data/chart1_default_by_age.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 1 saved successfully!")
```

```python
plt.figure(figsize=(10,5))
income_data = df[df['MonthlyIncome'] < 50000]['MonthlyIncome']

plt.hist(income_data, bins=50, color='#1D9E75', edgecolor='white', linewidth=0.3)

plt.title('Monthly Income Distribution', fontsize=14, fontweight='bold', pad=15)
plt.xlabel('Monthly Income(USD)', fontsize=12)
plt.ylabel('Number Of Customers', fontsize=12)
plt.axvline(income_data.median(), color='#E24B4A', linewidth=2,
            linestyle='--', label=f'Median: ${income_data.median():,.0f}')
plt.legend(fontsize=11)
plt.tight_layout()
plt.savefig('../data/chart2_income_distribution.png', dpi=150, bbox_inches='tight')
plt.show()
print("Chart 2 saved successfully!")
```
